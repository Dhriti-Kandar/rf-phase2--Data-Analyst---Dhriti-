# Business Questions

## Q1: Movies vs TV Shows Distribution

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Movies v/s TV shows")
plt.ylabel("count")
plt.show()
```

## Q2: Top 10 Content Producing Countries

```python
df['country'].value_counts().head(10)
```

##  Q3: Growth of Content Over Years

```python
content_growth = df['release_year'].value_counts().sort_index()
content_growth
```

```python
content_growth.plot(kind='line')
plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

Result: Massive growth between 2016–2019.

## Q4: Most Common Ratings

```python
df['rating'].value_counts()
```

## Q5: Most Popular Genres

```python
df['listed_in'].value_counts().head(10)
top_genres = df['listed_in'].value_counts().head(10)
top_genres
```

```python
plt.figure(figsize=(10,6))
top_genres.plot(kind='bar')

plt.title("Top 10 Most Common Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

### to conclude
-Netflix has more Movies than TV Shows.
-USA is the leading content contributor.
-Content growth increased significantly between 2016–2019.
-TV-MA and TV-14 are most common ratings.
-Drama and International genres dominate the platform.
